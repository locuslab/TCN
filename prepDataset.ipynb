```python
def custEnc(data):
    # define input string
    # define universe of possible input values
    alphabet = '0123456789abcdefghijklmnopqrstuvwxyz,._'
    # define a mapping of chars to integers
    char_to_int = dict((c, i) for i, c in enumerate(alphabet))
    int_to_char = dict((i, c) for i, c in enumerate(alphabet))
    # integer encode input data
    integer_encoded = [char_to_int[char] for char in data]
    #print(integer_encoded)
    # one hot encode
    onehot_encoded = list()
    for value in integer_encoded:
        letter = [0 for _ in range(len(alphabet))]
        letter[value] = 1
        onehot_encoded.append(letter)
    #print(onehot_encoded) # the real encoding
    #invert encoding
    #inverted = int_to_char[argmax(onehot_encoded[0])]
    #print(inverted)
    #print(type(onehot_encoded))
    return onehot_encoded
```

```python
from numpy import argmax
import numpy as np
# Using readlines() 
file1 = open('dsDropped.txt', 'r') 
Lines = file1.readlines() 
  
dataset = []
# Strips the newline character 
for line in Lines: 
    line1=line[:-1] # to get rid of the carriage return
    lineDataLower = line1.lower()
    lineData = custEnc(lineDataLower)
    dataset.append(lineData)

data = np.array(dataset, dtype = object)
print(data[0])
```
